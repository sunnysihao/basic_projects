```python
import json
import numpy as np
```

```python
with open(r"C:\Users\EDY\Downloads\pilot_Data-20221108T101954Z-001\pilot_Data\reconstructed_map\reconstructed_map.json", 'r') as f:
    j = f.read()
    json_content = json.loads(j)
```

```python
p = json_content['points']
p
```

```python
p[0]
```

```python
x = []
y = []
z = []
rgb_l = []
for obj in p:
    x.append(float(obj['x']))
    y.append(float(obj['y']))
    z.append(float(obj['z']))
    r = int(obj['r'])
    g = int(obj['g'])
    b = int(obj['b'])
    rgb = int((r << 16) + (g << 8) + b)
    rgb_l.append(rgb)
```

```python
x = np.array(x, dtype=np.float32)
y = np.array(y, dtype=np.float32)
z = np.array(z, dtype=np.float32)
rgb_l = np.array(rgb_l, dtype=np.uint32)
points = np.stack([x, y, z, rgb_l], axis=1)
```

```python
points
```

```python
save_pcd_file = r"C:\Users\EDY\Downloads\pilot_Data-20221108T101954Z-001\pilot_Data\reconstructed_map\01.pcd"
with open(save_pcd_file, 'w', encoding='ascii') as pcd_file:
    point_num = points.shape[0]
    heads = [
        '# .PCD v0.7 - Point Cloud Data file format',
        'VERSION 0.7',
        'FIELDS x y z rgb',
        'SIZE 4 4 4 4',
        'TYPE F F F U',
        'COUNT 1 1 1 1',
        f'WIDTH {point_num}',
        'HEIGHT 1',
        'VIEWPOINT 0 0 0 1 0 0 0',
        f'POINTS {point_num}',
        'DATA ascii'
    ]
    pcd_file.write('\n'.join(heads))
    for i in range(point_num):
        string_point = '\n' + str(points[i, 0]) + ' ' + str(points[i, 1]) + ' ' + str(points[i, 2]) + ' ' + str(points[i, 3])
        pcd_file.write(string_point)
```
